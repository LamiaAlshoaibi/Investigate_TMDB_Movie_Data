

### Data Cleaning 
-Dropping the column that i'm not going to use 

```python
movies.drop(['id','imdb_id','homepage','cast','keywords','tagline','overview','production_companies','budget_adj','revenue_adj'],axis=1,inplace=True)
movies.head(2)
```

-checking for null values and Dropping them

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

-check if any of columns have null values

```python
movies.isnull().sum().any()
```

-Dropping any duplicates:

```python
movies.drop_duplicates(inplace=True)
```

-converting the data type of release_date column from string to date

```python
movies.release_date=pd.to_datetime(movies.release_date)
```

check of all what we did 

```python
movies.info()
```

## Exploratory Data Analysis



### Research Question 1 (Which generes are most popular from year to year?)

-first we use the groupby function to see the most popular genres for each year  

```python
y=movies.groupby(['release_year'])['genres','popularity'].max()
y
```

then we look for the duplicated value in genres column and count them to find the most popular genres

```python
mostpopular=y.genres[y['genres'].duplicated()].value_counts()
mostpopular
```

it seems like the 'western' genres is the most popular one 

we can see that in the pie plot 

```python
mostpopular.plot(kind='pie',figsize=(8,8),title='Most popular movies genres');
```

we can see from the plot that the most popular movie genres from year to year is [western]

### Research Question 2  (What kind of properties are associated with movies that have high revenues?)

finding the movies that have revenue more than the mean of revenue. then using describe() for the properties which are associated with high revenue movies.

```python
highrevenue=movies.query('revenue>=revenue.mean()')
highrevenue.describe()
```

now we can see histograms of high revenue movies properties

```python
highrevenue.hist(figsize=(8,8));
```

we can see from the histogram above that popularity is high , and the budget skewed to the right , and we can tell that the release year is skewed to the left, which means the most high revenue movies was released in the last years .

#### lets define scatter plot function to answer the next two questions

```python
def scatterplot(x,y,title,xlabel,ylabel):
    plt.scatter(x =x, y = y,alpha=0.7)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel);
```

###  Research Question 3 (Is there any correlation between the budget and revenue?)

```python
scatterplot(movies.budget,movies.revenue,'correlation between budget and revenue','budget','revenue')
```

we can say that There is positive correlation between budget and revenue

###  Research Question 4 (Is there any correlation between the vote count and popularity?)

```python
scatterplot(movies.vote_count,movies.popularity,'correlation between vote count and popularity','vote count','popularity')
```

it seems like there is positive correlation between vote count and popularity.

### Research Question 5 (Data of Action Movies? Data of western movie? Bar plot of the Average revenue of the action and western movie)

function returns all the movie from specific genres

```python
def movie_genres(m_genres):
    m=movies[movies.genres==m_genres]
    return m
```

```python
movie_genres('Action')
```

```python
movie_genres('Western')
```

we use variables to take the action movies and western movies

```python
action=movie_genres('Action')
```

```python
western=movie_genres('Western')
```

mean revenue of Action and western movies genres bar plot

```python
action_mean_revenue=action.revenue.mean()
western_mean_revenue=western.revenue.mean()
locations=[1,2]
heights=[action_mean_revenue,western_mean_revenue]
labels=['Action','western']
colors=['blue','green']
plt.bar(locations,heights,tick_label=labels,color=colors)
plt.title('Average revenue of Action and western movies')
plt.xlabel('movie genres')
plt.ylabel('revenue');
```

### Research Question 6 (What is the title of the most popular movie ?)

use most_popular variable to take the maximum popularity , then finding the movie title where the popularity equals the maximum 

```python
most_popular=movies.popularity.max()
most_popular_movie=movies.original_title[movies.popularity==most_popular]
most_popular_movie 
```
